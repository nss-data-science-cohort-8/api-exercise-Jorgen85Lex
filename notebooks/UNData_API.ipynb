```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
import json
import numpy as np
```

```python
endpoint_GDP = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'

params_GDP = {
'per_page': 20000
}

response_GDP = requests.get(endpoint_GDP,params = params_GDP)
response_GDP
```

```python
GDP_df = pd.json_normalize(response_GDP.json()[1])
GDP_df
```

```python
endpoint_LE = 'https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json'

params_LE = {
'per_page': 20000
}

response_LE = requests.get(endpoint_LE, params=params_LE)
response_LE
```

```python
life_expectancy_df = pd.json_normalize(response_LE.json()[1])
life_expectancy_df
```

```python
GDP_df = GDP_df.rename(columns=
              {'countryiso3code' : 'Country_Code',  
               'country.id' : 'Country_ID', 
               'country.value' : 'Country_Value',
               'date' : 'Date',
               'value' : 'GDP_Per_Capita'})
```

```python
GDP_df = GDP_df.drop(['unit', 'obs_status', 'decimal','indicator.id', 'indicator.value'], axis=1)
GDP_df
```

```python
life_expectancy_df = life_expectancy_df.rename(columns=
              {'countryiso3code' : 'Country_Code',
               'country.id' : 'Country_ID', 
               'country.value' : 'Country_Value',
               'date' : 'Date',
               'value' : 'Life_Expectancy'})
```

```python
life_expectancy_df = life_expectancy_df.drop(['unit', 'obs_status', 'decimal', 'indicator.id', 'indicator.value'], axis=1)
life_expectancy_df
```

```python
merged_GDP_le = pd.merge(GDP_df, life_expectancy_df, how="inner", on = ["Country_Code", "Date", "Country_ID", "Country_Value"])
merged_GDP_le.head()
```

```python
endpoint_country = 'https://api.worldbank.org/v2/country/all?format=json'

params_country = {
'per_page': 300
}

response_country = requests.get(endpoint_country, params = params_country)
response_country
```

```python
country_df = pd.json_normalize(response_country.json()[1])
country_df.head()
```

```python
country_df = country_df.rename(columns=
              {'name': 'Country_Value',
               'id': 'Country_Code',
              'iso2Code': 'Country_ID'}
)
country_df
```

```python
merged_GDP_le_country = pd.merge(merged_GDP_le, country_df, how="inner", on= ["Country_Code", "Country_ID", "Country_Value"])
merged_GDP_le_country
```
